```python
x = ['Jan', 'Feb', 'March', 'April', 'May']
y1 = np.array([75, 144, 114, 102, 108])
y2 = np.array([90, 126, 102, 84, 126])
y3 = np.array([96, 114, 75, 105, 135])
y4 = np.array([105, 90, 150, 90, 75])
y5 = np.array([90, 75, 135, 75, 90])
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize = (12,7))
plt.bar(x, y1, color='slategrey')
plt.bar(x, y2, bottom=y1, color='lightseagreen')
plt.bar(x, y3, bottom=y1+y2, color='lightsteelblue')
plt.bar(x, y4, bottom=y1+y2+y3, color='grey')
plt.bar(x, y5, bottom=y1+y2+y3+y4, color='royalblue')
plt.xlabel("Months")
plt.ylabel("Traffic Sources")
plt.legend(["Organic", "Paid", "Referral", "Direct","Social"],loc = "best")
plt.title("Traffic Sources by Month")
plt.show()
```
