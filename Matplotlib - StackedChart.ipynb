```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.available
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
Apr = [5, 7, 6, 8, 7, 9]
May = [0, 4, 3, 7, 8, 9]
June = [6, 7, 4, 5, 6, 8]
Jul = [7, 2, 5, 2,6, 6]
```

```python
labels = ["April ", "May", "June","Jul"]
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(15,7))
ax.stackplot(x,Apr,May,June,Jul,labels = labels) 
ax.legend(loc=2)
```
