```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.available
```

```python
stock = pd.read_csv('GOOG.csv')
stock
```

```python
indexed_stock = stock.set_index('Date')
```

```python
ts = indexed_stock[['High']]
```

```python
ts
```

```python
plt.style.use('seaborn')
plt.plot(ts)
```

```python
plt.show()
```

```python
ts_low = indexed_stock[['Low']]
```

```python
ts_low
```

```python
fig, ax = plt.subplots()
ax.plot(ts, c = 'blue')
ax.plot(ts_low, c = 'black')
plt.show()
```
